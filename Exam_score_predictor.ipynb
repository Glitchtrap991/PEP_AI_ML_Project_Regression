### Removing Null Values

The `df.dropna()` method can be used to remove rows or columns with missing values. By default, it removes rows that contain *any* `NaN` values. You can specify different parameters to control its behavior.

````python
# To remove rows that contain ANY null values:
df_no_null_rows = df.dropna()
print("DataFrame after dropping rows with any null values:")
display(df_no_null_rows.head())

# To remove columns that contain ANY null values:
df_no_null_cols = df.dropna(axis=1)
print("\nDataFrame after dropping columns with any null values:")
display(df_no_null_cols.head())

# To remove rows only if ALL values in that row are null:
df_all_null_rows = df.dropna(how='all')
print("\nDataFrame after dropping rows where ALL values are null:")
display(df_all_null_rows.head())
````

Alternatively, if you want to replace null values instead of removing them, you can use the `df.fillna()` method. For example, to fill all nulls with 0:

```python
df_filled = df.fillna(0)
```

Or with the mean of the column:

```python
df['column_name'] = df['column_name'].fillna(df['column_name'].mean())
```

````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='study_hours', y='exam_score', data=df, scatter_kws={'alpha':0.3}, line_kws={'color': 'red', 'linewidth': 3})
plt.title('Regression Plot of Exam Score vs. Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust x and y axis limits to zoom out
plt.xlim(df['study_hours'].min() - 0.5, df['study_hours'].max() + 0.5)
plt.ylim(df['exam_score'].min() - 5, df['exam_score'].max() + 5)

plt.show()
````
